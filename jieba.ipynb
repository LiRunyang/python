```python
name=['高加林','刘巧珍','黄亚萍','张克南','德顺爷','刘立本','高玉德','高明楼']
jieba.add_word('高加林')
jieba.add_word('张克南')
jieba.add_word('德顺爷')
jieba.add_word('高明楼')
import jieba
with open('rensheng.txt',encoding='utf-8') as f:
    file=f.read()
a=file.split('\n\u3000\u3000')
def countnum(t):
    l1=jieba.lcut(t)
    b=[]
    for i in name:
        b.append(l1.count(i))
    for i in l1:
        if i=='加林':
            b[0]+=1
        elif i=='巧珍':
            b[1]+=1
        elif i=='德顺':
            b[5]+=1
    c=[]
    a=-1
    for i in b:
        a+=1
        if i!=0:
            c.append(name[a])
            #print(name[a])
    return c
para_num=[]
for i in a:
    para_num.append((countnum(i)))
for i in para_num:
    if i==[]:
        para_num.remove(i)
for i in para_num:
    print(i)
```
