```python
#例
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,360)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
#例二
import matplotlib.pyplot as plt
y16=[13431,13241,13241,13411,12341,12341]
x16=[0.8,1.8,2.8,3.8,4.8,5.8]
y17=[13244,13412,12344,31433,13411,12344]
x17=[1.2,2.2,3.2,4.2,5.2,6.2]
x=[1,2,3,4,5,6]
xname=['GD','JS','SD','ZJ','HN','SC']
plt.bar(x16,y16,width=0.3,label='2016')
plt.bar(x17,y17,width=0.3,label='2017')
plt.legend()
plt.title('GDP')
plt.xticks(x,xname)
plt.show()
```
