## 1.函数的定义

```python
def func_jiecheng(n):
    s=1
    for i in range(1,n+1):
        s=s*i
    return s

print(func_jiecheng(7))
```

* 2.1 按<font color=red size=3>位置</font>传递

```python
#%load c:\python37\z3.3.py
from math import *
def func_norm(u,a,x):
    fx=1/(sqrt(2*pi)*a)*exp(-1*(x-u)**2/(2*a**2))
    return fx
```

```python
mean=eval(input('输入均值：'))
sdv=eval(input('输入标准差：'))
x=eval(input('输入x：'))
print(func_norm(mean,sdv,x))
```

```python
import matplotlib.pyplot as plt
x=[]
y=[]
i=0
while i<=200:
    x.append(i)
    i+=0.1
    y.append(func_norm(100,10,i))
plt.plot(x,y)
plt.show()  
```

* 2.2 可变数量参数<br>
格式：def func(a,\*b),<br>
其中的b在调用时被视为一可变长度的元组

```python
def func_max(a,*b):
    m=a
    for i in b:
        if i>m:
            m=i
    return m

print(func_max(7,9,10,1,3,5))    
```

## 4.递归
* 在函数的定义中调用函数自身    
* 基例（一个或多个，比如数列）；递归链（第n次要用到第n-1次）

【例】猴子吃桃问题如下：猴子第一天摘下若干个桃子，当即吃了一半，还不过瘾，又多吃了⼀个，第二天早上又将剩下的桃⼦吃掉一半，又多吃了一个。以后每天早上都吃前一天剩下的一半再加一个。到第n天早上想再吃时，见只剩下⼀个桃⼦了。求第一天共摘多少个桃子？ 用递归方法实现求桃⼦数⽬的函数，其输⼊参数包括n，返回值为桃子个数。在主程序中，首先输⼊n的值，然后调用该函数并打印得到的结果。

```python
def func_peach(n):
    if n==1:
        peach=1
    else:
        peach=2*func_peach(n-1)+1
    return peach

n=100
print('第一天摘桃子=',func_peach(n))
```

默认递归深度为1000

【例】字符串反转，比如输入的是'1234567'，输出为'7654321'

```python
def func_reverse(s):
    if len(s)==1:
        s_r=s
    else:
        s_r=func_reverse(s[1:])+s[0]
    return s_r

print(func_reverse("ABCD"))
```

【例】斐波那契数列

```python
def func_feibo(n):
    if n==1:
        feibo=0
    elif n==2:
        feibo=1
    else:
        feibo=func_feibo(n-1)+func_feibo(n-2)
    return(feibo)

n=20  #30以上速度将很慢！！
for i in range(1,n+1):
    print(func_feibo(i),end='  ')    
```
